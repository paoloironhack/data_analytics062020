```python
#create function and apply to original dataframe
def country_converter(x):
    '''
    Input: A country code
    Output: Find country code in Legenda CSV and return the country name
    
    Example
    input: 'T001047'
    output: 'Totaal landen'
    '''
    return legenda[legenda.ID == x].Title.values[0]
```

```python
#change country codes to country names
hoteldata['WoonlandVanGasten'] = hoteldata['WoonlandVanGasten'].apply(country_converter)

#example code of applying function to column
#df['z'] = df['z'].apply(function)
```

```python
#change region codes to city
hoteldata['RegioS'] = hoteldata['RegioS'].apply(country_converter)
```

```python
#create function to change all data formats to year-month
def change_dateformat (text):
    '''
    Input: data format with MM
    Output: data format with -
    
    Example
    input: '2018MM10'
    output: '2018-10'
    '''
    return re.sub('MM', '-', text)
```

```python
#change data format in dataframe
hoteldata['Perioden'] = hoteldata['Perioden'].apply(change_dateformat)
hoteldata
```

```python
#add -01 to each date
hoteldata.Perioden = hoteldata.Perioden.apply(lambda x: x + '-01')
hoteldata

#example code
#data.Col1 = data.Col1.apply(lambda x: x + '_'*(max_length - len(x)))
```
