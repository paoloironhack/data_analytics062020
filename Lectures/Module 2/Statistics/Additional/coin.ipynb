```
from random import randint
from collections import Counter
```

```
num = 10000
flips = [randint(0,1) for r in range(num)]
results = []
for object in flips:
        if object == 0:
            results.append('Heads')
        elif object == 1:
            results.append('Tails')

print(results)
```

```
Counter(results)
```
