```python
def get_user_input():
    year_start = int(input('Enter the from year (minimum 1984):'))
    year_end   = int(input('Enter the to year (maximum 2012):'))
    #@TODO spice it up with year validation
    return (year_start, year_end)
#test:
'''
a,b = get_user_input()
print(a)
print(b)
'''
```

```python
def acquire(path_and_filename):
    df = pd.read_csv(path_and_filename)
    return df
```

```python
# WRANGLE:
# wrangle takes a dataframe and a year
# wrangnle outputs the same dataframe 
# containing only the records of the specified year
# 
def wrangle(df, user_year):
    filter_idx = (df.Year == user_year)
    df = df[filter_idx]
    return df
```

```python
# return top fuel eff by make
def analyze(df123):
    df123 = df123.groupby('Make').agg({'Combined MPG':'mean'}).reset_index()
    df123 = df123.sort_values(by='Combined MPG', ascending=False).head(10)
    return df123
```

```python
def report(df, year):
    fig, ax = plt.subplots(figsize=(15,8))
    barchart = sns.barplot(data=df, x='Make',y='Combined MPG')
    plt.title(f'Top 10 fuel efficient Makes in {year}', fontsize=16)
    return barchart
```

```python
def save_report(my_barchart, year):
    fig = my_barchart.get_figure()
    today = datetime.date(datetime.now())
    fig.savefig(f'export/Top10_{year}_at_{today}.png')
```

```python
# MAIN PROGRAM

df                    = acquire('vehicles.csv')

(year_start,year_end) = get_user_input()

for year in range(year_start, year_end+1):
    print('reporting and saving year :', year)
    df_wrangled           = wrangle(df, year)
    df_analysed           = analyze(df_wrangled)
    my_barchart           = report(df_analysed, year)
    save_report(my_barchart, year)
```
