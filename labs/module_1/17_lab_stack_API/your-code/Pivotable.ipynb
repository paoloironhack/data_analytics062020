###Pivot Tables#

```python
import pandas as pd
data_url = 'http://bit.ly/2cLzoxH'
df = pd.read_csv(data_url)
```

```python
df.head()
```

```python
df.sample(5)
```

```python
df.describe()
```

```python
#when selecting several columns, you must have a list of list 
df2=df[['continent','lifeExp']]
df2.sample(3)
```

```python
df2
```

```python
pd.pivot_table(df,values='lifeExp',columns='continent')
```

```python
pd.pivot_table(df,values='lifeExp',columns='country').transpose()
```

```python
df3=df[['continent','year','lifeExp']]
df.sample(3)
```

```python
pd.pivot_table(df,values='lifeExp',
                    index=['year'],
                      columns='continent')
```

```python
pd.pivot_table(df,values='lifeExp',
                    index=['year'],
                      columns='continent',
                         aggfunc=['min','max'])
```
