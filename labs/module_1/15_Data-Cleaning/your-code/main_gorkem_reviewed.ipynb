# Data Cleaning 

#### 1. Import pandas library.

```python
import pandas as pd
import numpy as np
```

#### 2. Import pymysql and sqlalchemy as you have learnt in the lesson of importing/exporting data. 


```python
import pymysql
from sqlalchemy import create_engine
```

#### 3. Create a mysql engine to set the connection to the server. 
Check the connection details here [here](https://relational.fit.cvut.cz/dataset/Stats)

```python
engine = create_engine('mysql+pymysql://guest:relational@relational.fit.cvut.cz/stats')
```

#### 4. Import the users table.

```python
users_table = pd.read_sql_query("SELECT * FROM stats.users",engine)
users_table
```

#### 5. Rename Id column to userId.

```python
users_table = users_table.rename(columns = {"Id" : "userId"})

users_table.head()

#[Tjerk]: yes!
```

#### 6. Import the posts table. 

```python
posts_table = pd.read_sql_query("SELECT * FROM stats.posts",engine)

#[Tjerk]: yes! and use a space after your comma for consistency! 
```

#### 7. Rename Id column to postId and OwnerUserId to userId.

```python
posts_table = posts_table.rename(columns = {"Id" : "postId", "OwnerUserId" : "userId"})
posts_table.head()
#[Tjerk]: yes!
```

#### 8. Define new dataframes for users and posts with the following selected columns:
**users columns**: userId, Reputation, Views, UpVotes, DownVotes  
**posts columns**: postId, Score, userID, ViewCount, CommentCount

```python
users_db = users_table[['userId',"Reputation","Views","UpVotes","DownVotes"]]
posts_db = posts_table[["postId","Score","userId","ViewCount","CommentCount"]]

#[Tjerk]: yes!
```

#### 9. Merge the new dataframes you have created, of users and posts. 
You will need to make an inner [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
merge_db = users_db.merge(posts_db, on = "userId")

#[Tjerk]: yes! and maybe it is also nice for next time to specify what kind of join, e.g. inner.
```

#### 10. How many missing values do you have in your merged dataframe? On which columns?

```python
merge_db.count()
m_value = []

for data in merge_db:
    m_value.append(merge_db[data].isnull().sum())
    print(data,"column has", m_value[-1] , "missing value(s) out of", len(merge_db["ViewCount"]) ,"records.\n")

    
#[Tjerk]: yes! nice, with a for loop! can you also do it without a for loop?
# and don't forget to answer the question that is asked.
# we are not so interested in the 0 missing values!
```
