### Getting data on ESG grading

```python
# Importing libraries
import yfinance as yf
import pandas as pd
from yahoofinancials import YahooFinancials
```

```python
# Looking at the sustainability scores one company
cola = "KO"
cola_y = yf.Ticker(cola)
esg_data = pd.DataFrame.transpose(cola_y.sustainability)
esg_data['company_ticker'] = str(cola_y.ticker)
#esg_data
```

```python
# Import list of 30 tickers from DOW 30
djia = pd.read_csv('./data/ticker.csv')
```

```python
ticker_lst = list(djia['ticker_code'])
#ticker_lst
```

```python
# Retrieve Yahoo! Finance sustainability scores for 
#each ticker and skip the ones tha does not have ESG info

for i in djia['ticker_code']:
    # print(i)
    i_y = yf.Ticker(i)
    try:
        if i_y.sustainability is not None:
            temp = pd.DataFrame.transpose(i_y.sustainability)
            temp['company_ticker'] = str(i_y.ticker)
            # print(temp)
            esg_data = esg_data.append(temp)
    except IndexError:
        pass

esg_data.head(3)
```
